```python
fig, axes = plt.subplots(1, 3, figsize=(15, 4))

class_counts.plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Documents per Class')
axes[0].set_ylabel('Count')
axes[0].tick_params(axis='x', rotation=45)

axes[1].hist(df['word_count'], bins=30, color='coral', edgecolor='black')
axes[1].set_title('Document Length Distribution')
axes[1].set_xlabel('Word Count')
axes[1].set_ylabel('Frequency')
axes[1].axvline(df['word_count'].median(), color='red', linestyle='--', label='Median')
axes[1].legend()

axes[2].hist(df['ocr_ratio'], bins=20, color='lightgreen', edgecolor='black')
axes[2].set_title('OCR Usage per Document')
axes[2].set_xlabel('OCR Ratio')
axes[2].set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
